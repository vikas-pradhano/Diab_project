```python
import pandas as pd
import numpy as np
df=pd.read_csv('https://raw.githubusercontent.com/shsarv/Machine-Learning-Projects/main/Diabetes%20Prediction%20%5BEND%202%20END%5D/dataset/kaggle_diabetes.csv')
```

```python
df
```

```python
df[df['BloodPressure']==0]
```

```python
df.loc[df['BloodPressure'] == 0, 'BloodPressure']=df['BloodPressure'].median()
```

```python
df[df['BloodPressure'] == 0]
```

```python
df
```

```python

df.loc[df['SkinThickness'] == 0, 'SkinThickness']=df['SkinThickness'].mean()
```

```python
df['SkinThickness']=df['SkinThickness']/10
```

```python
df
```

```python

df.loc[df['BMI'] == 0, 'BMI']=df['BMI'].mean()
```

```python
df.drop('Insulin',axis=1,inplace=True)
```

```python
df
```

```python
y=df['Outcome']
x=df.drop('Outcome',axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
lgr=LogisticRegression()
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
lgr.fit(x_test,y_test)
```

```python
lgr.score(x_test,y_test)
```

```python
lgr.score(x_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(lgr,x_train,y_train,cv=10)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score

print(confusion_matrix(y_test,lgr.predict(x_test)))
print(accuracy_score(y_test,lgr.predict(x_test)))
print(precision_score(y_test,lgr.predict(x_test)))
print(recall_score(y_test,lgr.predict(x_test)))
print(f1_score(y_test,lgr.predict(x_test)))
```
